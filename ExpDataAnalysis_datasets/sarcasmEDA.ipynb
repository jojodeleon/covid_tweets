## Dataframe of news headlines from either The Onion or HuffPost 
### Has article link, headline, and a binary column titled is_sarcastic, article line, and headline
#### is_sarcastic: 0= true news, 1=sarcastic

```python
import pandas as pd
```

```python
sarcasm1 = pd.read_json('Sarcasm_Headlines_Dataset.json', lines=True)
```

```python
sarcasm1
```

```python
sarcasm1.headline[26708]
```

```python
sarcasm2 = pd.read_json('Sarcasm_Headlines_Dataset_v2.json', lines=True)
```

```python
sarcasm2
```

```python
sarcasm2.headline[28616]
```

```python
all_sarc = pd.concat([sarcasm1, sarcasm2], ignore_index=True)
all_sarc
```

```python
sarc_headlines = pd.DataFrame([all_sarc.headline, all_sarc.is_sarcastic]).T
sarc_headlines
```

```python
sarc_headlines.rename(columns={'is_sarcastic':'real_or_fake'}, inplace=True)
sarc_headlines
```

```python
sarc_headlines.info()
```

```python
sarc_headlines.to_csv('sarc_headlines.csv', index=False)
```
